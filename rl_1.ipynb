как работает процесс принятия решений в терминах MDP: состояния, действия, вознаграждения, траектории и возврат (return).

Среда = Марковский процесс принятия решений (MDP)  
(S,A,P,R,γ)

Возврат (return):
G(t) = R(t+1) + γ * R(t+2) + γ^2 * R(t+3) + ...

симуляция случайной политики в среде FrozenLake-v1

```python
import gymnasium as gym
import numpy as np
```

```python
env = gym.make("FrozenLake-v1", is_slippery=False)
gamma = 0.9
episodes = 10
```

```python
for ep in range(episodes):
    state, _ = env.reset()
    done = False
    trajectory = []
    total_return = 0
    t = 0

    while not done:
        action = env.action_space.sample()  # random политика
        next_state, reward, terminated, truncated, _ = env.step(action)
        done = terminated or truncated
        trajectory.append((state, action, reward))
        total_return += (gamma ** t) * reward
        state = next_state
        t += 1

    print(f"эпизод {ep+1}: траектория = {trajectory}, возврат = {total_return:.2f}")
```
